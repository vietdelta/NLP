```python
from word2vec import Word2Vec
file_text = "./Vietnamese_FaceBook_October_processed.txt"
f = open(file_text,"r")
sentences = []
for line in f:
    if(len(line)>1):
        sentences.append(line)
```

```python
w2v = Word2Vec(epochs=10,window_size=3,lr=0.01)
w2v.generate_data(sentences)
```

```python
w2v.fit()
```
